## Simple Linear Regression Analysis

The regression analysis is a statistical method to deal with the formulation of mathematical model depicting relationship amongst variables, which can be used for the purpose of prediction of the values of dependent variable, given the values of independent variables.

In simple linear regression, we have only two variables:<br>
• Dependent variable (also called Response), usually denoted as Y.<br>
• Independent variable (alternatively called Regressor), usually denoted as x.<br>
• A reasonable form of a relationship between the Response Y and the Regressor x is the linear relationship, that is in the form Y = α + βx<br>
• The concept of regression analysis deal with finding the best relationship between Y and x (and hence best fitted values of α and β) quantifying the strength of that relationship. <br>

• The task in regression analysis is to estimate the regression coefficients α and β.
• Suppose, we denote the estimates a for α and b for β. Then the fitted regression line is  Y = a + bx.

```python
def b_val(df1):
    # m_x = Mean of X | m_y = Mean of Y
    # b_num = Numerator b_den = Denominator
    
    m_x = df1['BasePay'].mean()
    m_y = df1['Response'].mean()
    b_num = ((df1['BasePay']-m_x) * (df1['Response']-m_y)).sum()
    b_den = ((df1['BasePay']-m_x) ** 2).sum()
    b = b_num/b_den
    print("The Value of b in Linear is : ", b)
    return b
```

```python
def a_val(df1,b):
    x_m = df1['BasePay'].mean()
    y_m = df1['Response'].mean()
    a = y_m - b*x_m
    print("The Value of a in Linear is : ",a)
    return a
```

```python
def plot_regression_line(df1,a,b):
    y_pred = a + b*df1['BasePay']
    sns.scatterplot(x = 'BasePay', y = 'Response', data = df1)
    plt.plot(df1['BasePay'],y_pred,color = 'red')
    plt.title('BasePay vs All together')
    return y_pred
```

### Least Square Method
• The residual sum of squares is often called the sum of squares of the errors about the fitted line and is denoted as SSE<br> 


• A quantity R^2, is called coefficient of determination is used to measure the proportion of variability of the fitted model.

R^2 = 1 - (SSE/SST)

• If fit is perfect, all residuals are zero and thus R^2 = 1.0 (very good fit)<br>
• If SSE is only slightly smaller than SST, then R^2 ≈ 0 (very poor fit)

```python
def R2_val(df1,y_pred_linear):
    m_y = df1['Response'].mean()
    sst = ((df1['Response'] - m_y) **2 ).sum()
    sse = ((df1['Response'] - y_pred_linear) **2 ).sum()
    R2 = 1 - (sse/sst)
    print("The Value of R^2 in Linear is : ", R2)
    return R2
```

```python
b = b_val(train)
```

```python
a = a_val(train,b)
```

```python
y_pred_linear = plot_regression_line(train,a,b)
```

```python
R2_Linear = R2_val(train,y_pred_linear)
```

## Non Linear Regression with degree 2

```python
def deg2_coeff(df1):
    x = df1['BasePay']
    y = df1['Response']
    A = np.array([
        [(x**4).sum(), (x**3).sum(),(x**2).sum()],
        [(x**3).sum(),(x**2).sum(),(x.sum())],
        [(x**2).sum(),(x.sum()),x.count()]
    ])
    B = np.array([
        ((x**2)*y).sum(),
        (x*y).sum(),
        y.sum()
    ])
    b_deg2 = np.linalg.solve(A, B)
    print("The Coefficients in Non Linear Degree 2 are : ",b_deg2)
    return b_deg2
```

```python
def plot_deg2_1(df1,b):
    x = df1['BasePay']
    y = df1['Response']
    y_pred2 = b[2] + b[1]*x + b[0]*x**2
    sns.scatterplot(x = x,y = y)
    plt.plot(x, y_pred2, 'go--', linewidth=1, markersize=1)
    plt.title('Non-linear Degree 2')
    return 
```

```python
def plot_deg2_2(df1, b):
    x = df1['BasePay']
    y = df1['Response']
    y_pred2 = b[2] + b[1]*x + b[0]*x**2
    plt.plot(x, y, 'o', color='blue', label='Actual')
    plt.plot(x, y_pred2, 'o', color='red', label='Predicted',linewidth=1)
    plt.xlabel('BasePay')
    plt.ylabel('All Together')
    plt.title('Non-linear Degree 2')
    plt.legend()
    plt.show()
    return y_pred2
```

```python
b_deg2 = deg2_coeff(train)
```

```python
y_pred_deg2_1 = plot_deg2_1(train,b_deg2)
```

```python
y_pred_deg2_2 = plot_deg2_2(train, b_deg2)
```

```python
R2_de2 = R2_val(train,y_pred_deg2_2)
```

## Non Linear with Degree 3

```python
def deg3_coeff(df1):
    x = df1['BasePay']
    y = df1['Response']
    A = np.array([
        [(x**6).sum(),(x**5).sum(),(x**4).sum(),(x**3).sum()],
        [(x**5).sum(),(x**4).sum(),(x**3).sum(),(x**2).sum()],
        [(x**4).sum(),(x**3).sum(),(x**2).sum(),x.sum()],
        [(x**3).sum(),(x**2).sum(),x.sum(),x.count()]
    ])
    B = np.array([
        ((x**3)*y).sum(),
        ((x**2)*y).sum(),
        (x*y).sum(),
        y.sum()
    ])
    b_deg3 = np.linalg.solve(A, B)
    print("The Coefficients in Non Linear Degree 3 are : ",b_deg3)
    return b_deg3
```

```python
def plot_deg3_1(df1,b):
    x = df1['BasePay']
    y = df1['Response']
    y_pred3 = b[3] + b[2]*x + b[1]*x**2 + b[0]*x**3
    plt.title('Non-linear Degree 3')
    sns.scatterplot(x = x,y = y)
    plt.plot(x, y_pred3, 'go--', linewidth=1, markersize=1)
    return y_pred3
```

```python
def plot_deg3_2(df1, b):
    x = df1['BasePay']
    y = df1['Response']
    y_pred3 = b[3] + b[2]*x**1 + b[1]*x**2 + b[0]*x**3
    
    plt.plot(x, y, 'o', color='blue', label='Actual')
    plt.plot(x, y_pred3, 'o', color='red', label='Predicted')
    plt.xlabel('BasePay')
    plt.ylabel('All Together')
    plt.title('Non-linear Degree 3')
    plt.legend()
    plt.show()
    return y_pred3
```

```python
b_deg3 = deg3_coeff(train)
```

```python
y_pred_deg3_1 = plot_deg3_1(train,b_deg3)
```

```python
y_pred_deg3_2 = plot_deg3_2(train, b_deg3)
```

```python
R2_de3 = R2_val(train,y_pred_deg3_1)
```

```python
R2_de3 = R2_val(train,y_pred_deg3_2)
```
